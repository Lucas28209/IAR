```python
import numpy as np
import random as rn
```

```python
direcoes = ['N','S','L','O']
```

```python
def direcao():
    return direcoes[rn.randint(0,3)]      
```

```python
posicao = np.random.randint(2, size=(10,10))
copia = posicao.copy() 
i = rn.randint(0,len(posicao)-1)
j = rn.randint(0,len(posicao)-1)

#anterior = posicao[i][j] 
#posicao[i][j] = 22

#fazzer uma copia da matrix posicao 
#posicao[i][j] = copia[i][j]

print(posicao)

dir = direcao()
for k in range(len(posicao)+10):
    aux = rn.randint(1,100)
   
    #print(dir,aux)

    if (aux%2 == 0):
        dir = direcao() 
    
    print(dir)
    print(i,j)      
    #anterior = posicao[i][j]
    if (dir == 'N' and i+1 < len(posicao)):    
        posicao[i][j] = copia[i][j]  
        i = i+1       
    elif (dir == 'S' and i-1 > 0):
        posicao[i][j] = copia[i][j]
        i=i-1        
    elif (dir == 'L' and j+1 < len(posicao)):
        posicao[i][j] = copia[i][j]
        j=j+1        
    elif (dir == 'O' and j-1 > 0):
        posicao[i][j] = copia[i][j]
        j=j-1        
    posicao[i][j] = 22

    print("\n",posicao)


```
